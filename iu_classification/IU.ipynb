```python
import cv2
import face_recognition
import os

# 저장할 폴더 지정
save_folder = 'face4'

# 폴더가 없으면 생성
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

cap = cv2.VideoCapture('./mp4data/AKMU.mp4')

# 시작 시간 설정 (20초부터)
start_time_in_seconds = 0
cap.set(cv2.CAP_PROP_POS_MSEC, start_time_in_seconds * 1000)

frame_count = 1  # 초기값 변경
max_frames = 50000
frames_between_saves = 1  # 10프레임마다 저장

while cap.isOpened() and frame_count < (1 + max_frames):
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % frames_between_saves == 0:
        face_locations = face_recognition.face_locations(frame)

        for top, right, bottom, left in face_locations:
            face_image = frame[top:bottom, left:right]
            resized_face_image = cv2.resize(face_image, (70, 70))
            cv2.imwrite(os.path.join(save_folder, f'faces_{frame_count}.jpg'), resized_face_image)

            frame_count += 1

cap.release()
```

```python
# 기존 폴더 지정
folder_path = './datasets/faces/IU_faces'

# 기존 폴더 내 파일들에 대한 처리
for idx, current_name in enumerate(sorted(os.listdir(folder_path))):
    # 기존 파일의 전체 경로
    current_path = os.path.join(folder_path, current_name)

    # 파일인지 확인
    if os.path.isfile(current_path):
        # 새로운 파일명 생성 (예: IU_1.jpg, IU_2.jpg, ...)
        new_name = f'IU_{idx + 1:04d}.jpg'

        # 새로운 파일의 전체 경로
        new_path = os.path.join(folder_path, new_name)

        # 파일 이름 변경
        os.rename(current_path, new_path)
        print(f'Renamed: {current_name} -> {new_name}')

print("Renaming complete.")

```

```python
# other_faces 에 있는 하위 폴더들을 others에 한곳에 other_2023부터 저장

import shutil

input_dir = 'other_faces1'
output_dir = 'others1'

# 모든 이미지를 저장할 폴더 생성
os.makedirs(output_dir, exist_ok=True)

# other_faces 폴더 안의 모든 하위 폴더에서 파일을 찾아서 other_2023.jpg 형식으로 변경하여 others 폴더에 저장
image_count = 0
for query_folder in os.listdir(input_dir):
    query_folder_path = os.path.join(input_dir, query_folder)
    if os.path.isdir(query_folder_path):
        for file_name in os.listdir(query_folder_path):
            source_path = os.path.join(query_folder_path, file_name)
            dest_path = os.path.join(output_dir, f'other_{image_count + 5533}.jpg')
            shutil.copy(source_path, dest_path)
            image_count += 1
```

```python
# others에 있는 사진들을 extracted_faces에 얼굴만 추출해서 70*70 사이즈로 저장

from PIL import Image

input_dir = './others1'
output_dir = './extracted_faces1'

# 모든 이미지를 저장할 폴더 생성
os.makedirs(output_dir, exist_ok=True)

# input_dir 폴더 안의 모든 이미지에 대해 얼굴을 추출하고 70x70 크기로 조정하여 저장
for file_name in os.listdir(input_dir):
    source_path = os.path.join(input_dir, file_name)
    
    # 이미지에서 얼굴 위치를 찾음
    image = face_recognition.load_image_file(source_path)
    face_locations = face_recognition.face_locations(image)
    
    # 얼굴이 발견되면 추출하여 70x70 크기로 조정하여 저장
    for i, (top, right, bottom, left) in enumerate(face_locations):
        face_image = image[top:bottom, left:right]
        pil_image = Image.fromarray(face_image)
        pil_image = pil_image.resize((70, 70), Image.BILINEAR)
        
        # 원래 파일과 같은 이름으로 저장
        dest_path = os.path.join(output_dir, file_name)
        pil_image.save(dest_path)

print("얼굴 추출 및 저장이 완료되었습니다.")
```
